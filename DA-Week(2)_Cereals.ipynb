```python
import pandas as pd

df = pd.read_excel("Cereals1.xls")  
df
```

```python
df['potass'].mean()
```

```python
df=df.mask(df==-1).fillna(df[['protein','fat','sodium','fiber','carbo','sugars','potass','vitamins']].mean())
df
```

```python
df.describe()
```

```python
Q1=df.quantile(q=0.25)
Q3=df.quantile(q=0.75)
IQR=Q3-Q1
print(IQR)
```

```python
df=df.mask(df>(1.5*IQR+Q3)).fillna(df[['protein','fat','sodium','fiber','carbo','sugars','potass','vitamins']].median())
df=df.mask(df<(Q1-1.5*IQR)).fillna(df[['protein','fat','sodium','fiber','carbo','sugars','potass','vitamins']].median())

df
```
