```python
import pandas as pd
import seaborn as sns
df = pd.read_csv("mtcars.csv")  
df
```

```python
#Find the car with the best mpg
print ("the car with best mpg is")
df['model'][df['mpg'].argmax()]
#or df1= df[['model']][df.mpg == df['mpg'].max()]
# print(df1)

```

```python
#Find the car with the worst mpg
print ("the car with worst mpg is")
df['model'][df['mpg'].argmin()]
```

```python
#Find the car with the best horsepower
print ("the car with best horsepower is")
df['model'][df['hp'].argmax()]
```

```python
#Find 5 number summary of displacement
print ('5 number summary of displacement')
df['disp'].describe()
```

```python
#Find median horse power
df['hp'].median()
```

```python
#What is average mpg for manual vs. automatic cars

print("avg mpg of automatic cars") 
index = df['am']==1
a=df.loc[index,'mpg'].mean()
print(a)

print("avg mpg of manual cars") 
index2 = df['am']==0
b=df.loc[index2,'mpg'].mean()
print(b)
```

```python
#Draw a histogram of miles per gallon
import seaborn as sns
sns.histplot(data=df,x='mpg')
```

```python
#Boxplot of mpg for each cylinder type
import seaborn as sns
sns.boxplot(data=df,y='mpg',x='cyl')
```

```python
#Create a crosstab displaying count of automatic vs. manual cars
pd.crosstab(df['am'],columns='count')
```

```python
#Create a crosstab displaying count of “am vs cyl”
pd.crosstab(df['am'],df['cyl'])
```

```python
#What is the correlation between the weight of the car and mpg
import seaborn as sns
sns.scatterplot(data=df,x='mpg',y='wt')
```
