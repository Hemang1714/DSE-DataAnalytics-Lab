### Import Statements and DataFrame Initialization 

```python
import pandas as pd
import numpy as np
import mlxtend
```

```python
df=pd.read_csv("groceries.csv")
pd.set_option('display.max_columns',None,'display.max_rows',None)
```

```python
df.head()
df.columns=range(0,33)
df.head()
```

### 1. How many transactions and items are there in the data set?

```python
Transactions=df.shape[0]
Transactions
```

```python
d_a=np.array(df).reshape(-1).astype(str)
```

```python
d_a=np.array([x for x in d_a if x.strip()!='nan'])
len(set(d_a))
```

### 2. Prepare the data for finding association rules. Each transaction will contain a list of item in the transaction.

```python
arr=np.array(df).astype(str)
```

```python
l=[]
for i in range(len(arr)):
    l.append([x for x in arr[i] if x!='nan'])
    
```

```python
l
```

### 3. Use Python library mlxtend and convert the transactions into a format that can be used in the Apriori method for finding frequent itemsets.

```python
from mlxtend.preprocessing import TransactionEncoder 
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
te=TransactionEncoder()
te.fit(l)
```

```python
ap=te.transform(l)
```

```python
items=list(set(d_a))
items.sort()
ap=pd.DataFrame(ap,columns=np.array(items))
```

```python
items[31]
```

```python
ap.head()
```

### 4. Find top selling items with minimum support of 2%.

```python
apriori(ap,min_support=0.02,use_colnames=True)
```

### 5. Find all frequent itemsets with minimum support of 5%

```python
apriori(ap,min_support=0.05,use_colnames=True)
```

### 6. Find all frequent itemsets of length 2 with minimum support of 2%.

```python
apri=apriori(ap,min_support=0.02,use_colnames=True,max_len=2)
```

```python
apri.loc[apri.itemsets.str.len()==2]
```

### 7. Find the top 10 association rules with minimum support of 2%, sorted by confidence in descending order.

```python
apri2=apriori(ap,min_support=0.02,use_colnames=True)
```

```python
association_rules(apri2,min_threshold=0.4).sort_values(by='confidence',ascending=False)[:10]
```

### 8. Find association rules with minimum support of 2% and lift of more than 1.0.

```python
association_rules(apri2,metric="lift",min_threshold=1).sort_values(by='lift',ascending=False)
```
